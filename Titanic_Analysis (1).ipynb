

The Titanic dataset contains passenger information and survival status.
The dataset has both numerical and categorical features.
Some columns such as Age, Cabin, and Embarked contain missing values.
Cabin has the highest number of missing values.
This dataset will require preprocessing before building a machine learning model.

handle missing values

```python

df['Age'].fillna(df['Age'].median(), inplace=True)


df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)


df.drop('Cabin', axis=1, inplace=True)


print(df.isnull().sum())
```


Age was filled using the median because it is a numerical feature.
Embarked was filled using the most frequent value (mode).
Cabin was removed because most of its values were missing, making it less useful.

import visualization libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

histogram

```python
plt.figure(figsize=(6,4))
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

boxplot (outliers)

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['Fare'])
plt.title("Fare Boxplot")
plt.show()
```

bar chart

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

correlation heatmap

```python
plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```
